```python
def f(x):
    if (x > 4) and (x <= 12):
        return 1
    elif (x > 12) and (x <= 16):
        return 2
    elif (x > 16) and (x <= 21) :
        return 3
    else:
        return 4
    
    ''' 1 Morning
        2 Noon
        3 Evening
        4 Night'''
```
