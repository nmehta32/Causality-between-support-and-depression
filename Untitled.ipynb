# preprocess function

```python
import pandas as pd
df_3 = pd.read_parquet("happyish_tweets.paraquet")
```

```python
import re
df_3['text'] = df_3['text'].apply(lambda x: re.split('https:\/\/.*', str(x))[0]) #remove links
```

```python
df_3["text"] = df_3["text"].str.replace(r's*https?://S+ (s+|$)', ' ').str.strip()
```

```python
df_3["text"] = df_3["text"].apply(lambda x: re.sub('@[^\s]+','',str(x))) ## remove @user
```

```python
# df_3['text']
spec_chars = ["!",'"',"#","%","&","'","(",")",
              "*","+",",","-",".","/",":",";","<",
              "=",">","?","@","[","\\","]","^","_",
              "`","{","|","}","~","â€“",'\n','RT']
for char in spec_chars:
    df_3['text'] = df_3['text'].str.replace(char, ' ')
```

```python
import pandas as pd
df_3 = pd.read_parquet("happyish_tweets.paraquet")

df_3
```

```python
# df_3.drop(['created_at','sd','Support_per_tweet'],axis = 1, inplace = True)
```

```python
df_3['Score'] = (df_3['public_metrics.reply_count']+df_3['public_metrics.retweet_count']+1) / (df_3['public_metrics.like_count']+2)
```

```python
unique_list = list(df_3['author_id'].unique())   
```

```python
import random

df_result=pd.DataFrame(columns = ['author_id','Score'])
res = []
for i in unique_list:
    df_i = df_3[df_3['author_id'] == i]
    l = len(df_i)
    r = random.randrange(0, l)
    k=0
    df_i.reset_index(inplace = True)
    try:
        while True:
            if df_i['labels'][k] != 1 :
                k+=1
            else:
                break
    except:
        pass
    for j in range(r):
        k += df_i['labels'][j]
    k_aft = 0
    for pp in range(r,l):
        k_aft += df_i['labels'][pp] 
    Result_df = pd.Series([])
    score = (k+1)/(k_aft+2)
    l = [i,score]
    res.append(l)
```

```python
df_result = pd.DataFrame (res, columns = ['author_id','Score'])
```

```python
df_result
```

```python
print('the average treatment effect of the intervention on non depressed people is ', df_result['Score'].mean(axis=0))
print('in the range ', df_result['Score'].min(),'to ',df_result['Score'].max())
```
